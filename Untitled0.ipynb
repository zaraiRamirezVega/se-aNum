```python
#Importar librerías
import cv2
import mediapipe as mp
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from IPython.display import display, Javascript
from google.colab.output import eval_js
from base64 import b64decode
from io import BytesIO
```

```python
# Función para abrir la cámara en modo espejo y capturar imagen (Colab)
def take_photo(filename='photo.jpg', quality=0.8):
  js = Javascript('''
    async function takePhoto(quality) {
      const div = document.createElement('div');
      const capture = document.createElement('button');
      capture.textContent = '📷 Tomar foto';
      div.appendChild(capture);

      const video = document.createElement('video');
      video.style.display = 'block';
      video.style.transform = 'scaleX(-1)';  // <-- Modo espejo para el video
      div.appendChild(video);

      document.body.appendChild(div);

      const stream = await navigator.mediaDevices.getUserMedia({video: true});
      video.srcObject = stream;
      await video.play();

      // Esperar que se presione el botón
      await new Promise((resolve) => capture.onclick = resolve);

      const canvas = document.createElement('canvas');
      canvas.width = video.videoWidth;
      canvas.height = video.videoHeight;
      const ctx = canvas.getContext('2d');

      // Voltear la imagen capturada para que salga espejo también
      ctx.translate(canvas.width, 0);
      ctx.scale(-1, 1);

      ctx.drawImage(video, 0, 0, canvas.width, canvas.height);

      stream.getTracks().forEach(track => track.stop());
      div.remove();

      return canvas.toDataURL('image/jpeg', quality);
    }
  ''')
  display(js)
  data = eval_js('takePhoto({})'.format(quality))
  binary = b64decode(data.split(',')[1])
  with open(filename, 'wb') as f:
    f.write(binary)
  return filename
```

```python


# === FUNCIONES DE DETECCIÓN ===

def pulgar_estirado(hand_landmarks, handedness):
    thumb_tip = hand_landmarks.landmark[4]
    thumb_ip = hand_landmarks.landmark[3]
    thumb_mcp = hand_landmarks.landmark[2]

    # Para ambas manos, el pulgar estirado estará más lejos del MCP
    # Y más "arriba" (menor valor de Y) que el IP
    return (
        abs(thumb_tip.x - thumb_mcp.x) > 0.06 and  # Distancia horizontal
        thumb_tip.y < thumb_ip.y  # Pulgar está "arriba"
    )



def detectar_numero_uno(hand_landmarks):
    return (
        hand_landmarks.landmark[8].y < hand_landmarks.landmark[6].y and  # Índice estirado
        hand_landmarks.landmark[12].y > hand_landmarks.landmark[10].y and
        hand_landmarks.landmark[16].y > hand_landmarks.landmark[14].y and
        hand_landmarks.landmark[20].y > hand_landmarks.landmark[18].y
    )


def detectar_numero_dos(hand_landmarks):
    return (
        hand_landmarks.landmark[8].y < hand_landmarks.landmark[6].y and  # Índice
        hand_landmarks.landmark[12].y < hand_landmarks.landmark[10].y and  # Medio
        hand_landmarks.landmark[16].y > hand_landmarks.landmark[14].y and
        hand_landmarks.landmark[20].y > hand_landmarks.landmark[18].y
    )


def detectar_numero_tres(hand_landmarks, handedness):
    return (
        hand_landmarks.landmark[8].y < hand_landmarks.landmark[6].y and
        hand_landmarks.landmark[12].y < hand_landmarks.landmark[10].y and
        pulgar_estirado(hand_landmarks, handedness) and
        hand_landmarks.landmark[16].y > hand_landmarks.landmark[14].y and
        hand_landmarks.landmark[20].y > hand_landmarks.landmark[18].y
    )


def detectar_numero_cuatro(hand_landmarks, handedness):
    return (
        hand_landmarks.landmark[8].y < hand_landmarks.landmark[6].y and
        hand_landmarks.landmark[12].y < hand_landmarks.landmark[10].y and
        hand_landmarks.landmark[16].y < hand_landmarks.landmark[14].y and
        hand_landmarks.landmark[20].y < hand_landmarks.landmark[18].y and
        not pulgar_estirado(hand_landmarks, handedness)
    )


def detectar_numero_cinco(hand_landmarks, handedness):
    return (
        hand_landmarks.landmark[8].y < hand_landmarks.landmark[6].y and
        hand_landmarks.landmark[12].y < hand_landmarks.landmark[10].y and
        hand_landmarks.landmark[16].y < hand_landmarks.landmark[14].y and
        hand_landmarks.landmark[20].y < hand_landmarks.landmark[18].y and
        pulgar_estirado(hand_landmarks, handedness)
    )


def detectar_numero_seis(hand_landmarks):
    thumb_tip = hand_landmarks.landmark[4]
    pinky_tip = hand_landmarks.landmark[20]
    distancia = np.sqrt((thumb_tip.x - pinky_tip.x)**2 + (thumb_tip.y - pinky_tip.y)**2)
    return distancia < 0.05


def detectar_numero_siete(hand_landmarks):
    thumb_tip = hand_landmarks.landmark[4]
    ring_tip = hand_landmarks.landmark[16]
    distancia = np.sqrt((thumb_tip.x - ring_tip.x)**2 + (thumb_tip.y - ring_tip.y)**2)
    return distancia < 0.05


def detectar_numero_ocho(hand_landmarks):
    thumb_tip = hand_landmarks.landmark[4]
    middle_tip = hand_landmarks.landmark[12]
    distancia = np.sqrt((thumb_tip.x - middle_tip.x)**2 + (thumb_tip.y - middle_tip.y)**2)
    return distancia < 0.05


def detectar_numero_nueve(hand_landmarks):
    thumb_tip = hand_landmarks.landmark[4]
    index_tip = hand_landmarks.landmark[8]
    distancia = np.sqrt((thumb_tip.x - index_tip.x)**2 + (thumb_tip.y - index_tip.y)**2)
    return distancia < 0.05

def detectar_numero_diez(hand_landmarks, handedness):
    pulgar_arriba = pulgar_estirado(hand_landmarks, handedness)
    dedos_doblados = (
        hand_landmarks.landmark[8].y > hand_landmarks.landmark[6].y and  # Índice doblado
        hand_landmarks.landmark[12].y > hand_landmarks.landmark[10].y and  # Medio doblado
        hand_landmarks.landmark[16].y > hand_landmarks.landmark[14].y and  # Anular doblado
        hand_landmarks.landmark[20].y > hand_landmarks.landmark[18].y    # Meñique doblado
    )
    return pulgar_arriba and dedos_doblados



```

```python
# === CAPTURAR FOTO ===
print("📸 Toma una foto usando tu cámara...")
image_path = take_photo()
```

```python
# Mostrar imagen original
image = Image.open(image_path).convert("RGB")

plt.figure(figsize=(6,6))
plt.imshow(image)
plt.axis('off')
plt.title("Imagen original capturada (modo espejo)")
plt.show()
```

```python
# Preparar imagen para MediaPipe
image_np = np.array(image)
image_bgr = cv2.cvtColor(image_np, cv2.COLOR_RGB2BGR)

# === PROCESAMIENTO CON MEDIAPIPE ===
mp_hands = mp.solutions.hands
mp_drawing = mp.solutions.drawing_utils
```

```python
with mp_hands.Hands(
    static_image_mode=True,
    max_num_hands=2,
    min_detection_confidence=0.5) as hands:

    results = hands.process(cv2.cvtColor(image_bgr, cv2.COLOR_BGR2RGB))
    annotated_image = image_np.copy()

    if results.multi_hand_landmarks:
        for hand_landmarks, hand_handedness in zip(results.multi_hand_landmarks, results.multi_handedness):
            handedness_label = hand_handedness.classification[0].label  # 'Left' o 'Right'

            mp_drawing.draw_landmarks(
                image=annotated_image,
                landmark_list=hand_landmarks,
                connections=mp_hands.HAND_CONNECTIONS,
                landmark_drawing_spec=mp_drawing.DrawingSpec(color=(255, 255, 0), thickness=4, circle_radius=5),  # Amarillo
                connection_drawing_spec=mp_drawing.DrawingSpec(color=(255, 0, 255), thickness=3))  # Rosado

            # Clasificar el gesto
            if detectar_numero_uno(hand_landmarks):
                print("☝️ ¡Número 1 detectado!")
            elif detectar_numero_dos(hand_landmarks):
                print("✌️ ¡Número 2 detectado!")
            elif detectar_numero_tres(hand_landmarks, handedness_label):
                print("🤟 ¡Número 3 detectado!")
            elif detectar_numero_cuatro(hand_landmarks, handedness_label):
                print("🖖 ¡Número 4 detectado!")
            elif detectar_numero_cinco(hand_landmarks, handedness_label):
                print("🖐️ ¡Número 5 detectado!")
            elif detectar_numero_seis(hand_landmarks):
                print("👌 ¡Número 6 detectado!")
            elif detectar_numero_siete(hand_landmarks):
                print("🫰 ¡Número 7 detectado!")
            elif detectar_numero_ocho(hand_landmarks):
                print("👆 ¡Número 8 detectado!")
            elif detectar_numero_nueve(hand_landmarks):
                print("👉 ¡Número 9 detectado!")
            elif detectar_numero_diez(hand_landmarks, handedness_label):
                print("👍 ¡Número 10 detectado!")
            else:
                print("✋ Mano detectada, pero no coincide con ningún número del 1 al 10.")
    else:
        print("⚠️ No se detectó ninguna mano.")
```

```python
# Mostrar imagen anotada
plt.figure(figsize=(6,6))
plt.imshow(annotated_image)
plt.axis('off')
plt.title("✋ Resultado con puntos de la mano")
plt.show()
```
